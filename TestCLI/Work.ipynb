```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv('full_history.csv')

exp1 = df1[df1['experiment'] == 1]
exp2 = df1[df1['experiment'] == 2]
exp3 = df1[df1['experiment'] == 3]
exp4 = df1[df1['experiment'] == 4]
exp5 = df1[df1['experiment'] == 5]
exp6 = df1[df1['experiment'] == 6]
exp7 = df1[df1['experiment'] == 7]
exp8 = df1[df1['experiment'] == 8]
exp9 = df1[df1['experiment'] == 9]
```

```python
composites = []
loc_f1 = []
dmg_f1 = []
for num in range(1,10):
    df = df1[df1['experiment'] == num]
    composites.append(np.max(df['composite']))
    if df['loc_f1'].notna().any():
        loc_f1.append(np.max(df['loc_f1']))
    else:
        try:
            loc_f1.append(loc_f1[-1])
        except:
            loc_f1.append(np.nan)
    if df['dmg_f1'].notna().any():
        dmg_f1.append(np.max(df['dmg_f1']))
    else:
        try:
            dmg_f1.append(loc_f1[-1])
        except:
            dmg_f1.append(np.nan)

plt.plot(np.arange(1,10), composites, label='Composite Score', marker='o')
plt.plot(np.arange(1,10), loc_f1, label='Segmentation F1', marker='o')
plt.plot(np.arange(1,10), dmg_f1, label='Damage F1', marker='o')
plt.xlabel('Experiment #')
plt.ylabel('Score')
plt.title('Experiment Score Progression', fontsize=14)
plt.legend()

plt.savefig('ScoreProgession.pdf')
```
